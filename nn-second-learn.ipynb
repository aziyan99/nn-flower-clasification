```python
import numpy

def sigmoid(x):
    return 1/(1+numpy.exp(-x))

def cost(b):
    return (b - 4)**2
```

```python
x = numpy.arange(-10., 10., 0.2)
sig = sigmoid(x)
plt.plot(x,sig)
plt.show()
```

```python
f = numpy.arange(-1., 5, 0.10)

for d in f:
    print(d)

co = cost(f)
plt.plot(f, co)
plt.show()
```
